# Creative Insights and Storytelling

## Diferencias entre los tipos.

### Si analizamos los promedios:

El tipo 1, se podría decir que tiene un valor intermedio entre los tipos 0 y 2 con los atributos alcohol, magnesium, ash, Proanthocyanins y proline. Lo que más diferencia tiene el tipo 1 es en malic_acid, color_intensity, ash_alcalinity, Nonflavanoid_Phenols; es el que mayor promedio tiene. Luego, Hue, Total_Phenols, Flavanoids, OD280 son los que menos promedio tienen. Estos atributos que poseen mayor y menor promedio destacan ya que entre el tipo 0 y el tipo 2 no hay una diferencia tan marcada como la hay con los atributos antes mencionados.

Hay muy poca diferencia entre los tipos 0 y 2, ya que si comparamos la mayoría de sus atributos vemos que son muy cercanos sus números, pero los que marcan la diferencia entre ellos son los atributos alcohol y proline. El alcohol del tipo 0 tiene menor promedio que el tipo 2, que aunque no sea grande la diferencia, es un punto a destacar. Pero con el atributo proline, vemos que en promedio tiene la diferencia más amplia, donde el tipo 0 posee un promedio de 497 y el tipo 2 un promedio de 1110. En el resto de los atributos no hay una diferencia muy grande.

Con respecto al atributo ash, casi no hay diferencia entre los promedios de los tipos.

### Si analizamos los máximos y mínimos:

Si vemos los máximos y los mínimos, vamos a notar números interesantes, ya que los que tienen menor promedio en algunos atributos son los que tienen el valor más grande como máximo, pero claramente también tienen el mínimo como el valor más pequeño. Algunos de estos casos son:
* Malic_Acid
* Ash
* Ash_Alcanity

Con respecto al resto de los atributos, vemos que se sigue manteniendo la diferencia entre los tipos.




## Datos

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargamos el dataset antes creado
df = pd.read_csv(r'data//wine_with_clusters.csv')
df.head()
```

Separamos por cluster para sacar conclusiones y poder pensar un storytelling. Y luego vemos el describe de cada uno.

```python
df_0 = df[df['Kmeans_Clusters'] == 0]
df_1 = df[df['Kmeans_Clusters'] == 1]
df_2 = df[df['Kmeans_Clusters'] == 2]
```

```python
df_0.describe()
```

```python
df_1.describe()
```

```python
df_2.describe()
```

### Graficamos para que sea mas visible

```python
plt.figure(figsize=(30,10))

plt.subplot(1,3,1)
barplot = sns.barplot(df_0.describe().loc['mean'])
# Mostrar los valores numéricos en las barras
for index, value in enumerate(df_0.describe().loc['mean']):
    barplot.text(index, value + 8 , str(round(value,2)), ha='center', va='bottom', rotation_mode='anchor', fontsize = 12)

plt.xticks(rotation=90)
plt.title('Tipo 0', fontsize = 20, y = 1.01)

plt.subplot(1,3,2)
barplot = sns.barplot(df_1.describe().loc['mean'], color='orange')
# Mostrar los valores numéricos en las barras
for index, value in enumerate(df_1.describe().loc['mean']):
    barplot.text(index, value + 8 , str(round(value,2)), ha='center', va='bottom', rotation_mode='anchor', fontsize = 12)

plt.xticks(rotation=90)
plt.title('Tipo 1', fontsize = 20, y = 1.01)

plt.subplot(1,3,3)
barplot = sns.barplot(df_2.describe().loc['mean'], color='green')
# Mostrar los valores numéricos en las barras
for index, value in enumerate(df_2.describe().loc['mean']):
    barplot.text(index, value + 8 , str(round(value,2)), ha='center', va='bottom', rotation_mode='anchor', fontsize = 12)

plt.xticks(rotation=90)
plt.title('Tipo 2', fontsize = 20, y = 1.01)

plt.suptitle('Diferencia con los promedios de los tipos de vino', fontsize=30)
plt.show()
```
